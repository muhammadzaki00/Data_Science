# Menghapus elemen

```python
countries = ['Indonesia','Malaysia','Jepang','Brunei Darussalam']
print(countries)

# remove = hapus nama datanya
countries.remove('Malaysia')
print(countries)

# pop = menghapus alamatnya atau index nya
countries.pop(2) # 
print(countries)

# del
del countries[1]
print(countries)
```

# For Looping (Perulangan)

```python
for country in countries:
    print(country)
```

```python
for negara in countries:
    print(negara)

print("=========")

for negara in countries:
    if negara == 'Indonesia':
        print(negara)

print("=========")

for i, negara in enumerate(countries):
    print(i, negara)
```
